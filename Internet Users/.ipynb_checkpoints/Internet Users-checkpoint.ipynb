```python
import pandas as pd
data_frame = pd.read_csv('Final.csv')
data_frame
```

```python
df1 = pd.DataFrame(data_frame, columns=['Entity', 'Year', 'Internet Users(%)', 'No. of Internet Users', 'Total Population', 'Cellular Subscription', 'Broadband Subscription'])
df1.columns = [c.replace(' ', '_') for c in df1.columns]
df1.columns = [c.replace('(%)', '_percent') for c in df1.columns]

def get_total_population_str(internet_users, perc_internet_users):
    percentage = 1 if perc_internet_users == 0 else perc_internet_users
    total = (float(internet_users) * 100) / float(percentage)
    total = round((total / 1000000), 2)
    if total == 0.0:
        return 0
    else:
        #return str(total) + 'M'
        return total

df1['Total_Population'] = df1.apply(lambda x: get_total_population_str(x['No._of_Internet_Users'], x['Internet_Users_percent']), axis=1)
df1.sort_values(['Year', 'Total_Population'], ascending=[False, True])
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Internet_Users(%)', y='Year', data=df1)
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool 

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("('Internet_Users(%)', Year)", "(@Internet_Users_percent, @Year)"),
    ("Entity", "@Entity")]
)

p = figure(title='Internet Usage', x_axis_label='Internet_Users_percent', y_axis_label='Year', plot_width=1000, plot_height=1000, tools=[TOOLTIPS])
p.circle('Internet_Users_percent', 'Year', size=10, source=df1)
show(p)
```
